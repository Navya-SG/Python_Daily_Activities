```python
import pandas as pd
```

```python
dj=pd.read_json("new_airline_data_2024 (1).json")
dc=pd.read_csv("old_airline_data_2023 (1).csv")
```

```python
print(dj,dc)
```

```python
print(dj.isnull().sum())
```

```python
dj.columns=dj.columns.str.lower().str.replace(" ","_")
dj["month"]=dj["month"].str.capitalize()
```

```python
dj.iloc[:,1:]=dj.iloc[:,1:].fillna(dj.iloc[:,1:].mean())
```

```python
print(dj)
```

```python
dj.dropna(subset=["flights_operated","passengers_carried","target_passengers"],how='all',inplace=True)
print(dj)
```

```python
mon=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
dj[dj["month"].isin(mon)]
```

```python
dj.drop_duplicates(subset="month",inplace=True)
print(dj)
```

```python
dc.columns=dc.columns.str.lower().str.replace(" ","_")
dc["month"]=dc["month"].str.capitalize()
print(dc)
```

```python
new=pd.concat([dc,dj],ignore_index=True)
```

```python
print(new)
```

```python
print(dc.count)
```

```python
#new["year"]=pd.Series([2023 if i < 48 else 2024 for i in new.index ],index=new.index)
new["year"]=[2023 if i < len(dc) else 2024 for i in range(len(new))]
```

```python
print(new)
```

```python
new[new["month"].isin(mon)]
```

```python
new=new.sort_values(["year","month"])
print(new)
```

```python
new["month"]=pd.Categorical(new["month"],categories=mon,ordered=True)
print(new)
```

```python
new["pct_diff_target"]= ((new["passengers_carried"] - new["target_passengers"]) / new["target_passengers"]) * 100
print(new["pct_diff_target"])
```

```python
target_non_achievement = new[new["pct_diff_target"]<0][["year","month","pct_diff_target"]]
print("target_non_achievement")
```

```python
total_passengers_carried = new.groupby("year")["passengers_carried"].sum()
total_flights_operated = new.groupby("year")["flights_operated"].sum()
average_pct_diff_target = new.groupby("year")["pct_diff_target"].mean()
annual_summary = pd.concat([total_passengers_carried,total_flights_operated,average_pct_diff_target],ignore_index=True,axis=1)
```

```python
print(total_passengers_carried,total_flights_operated,average_pct_diff_target,annual_summary)
```

```python
best=new.loc[new["pct_diff_target"].idxmax()]
worst=new.loc[new["pct_diff_target"].idxmin()]
```

```python
print(best,worst)
print(new)
```

```python
new=new.to_csv("combined_airline_data.csv",index=False)
target_non_achievement = target_non_achievement.to_csv("target_non_achievement",index=False)
annual_summary = annual_summary.to_json("yearly_summary.json",index=False)
```
