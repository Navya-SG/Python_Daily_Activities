```python
import pandas as pd
import zipfile
```

```python
dj=pd.read_json("new_airline_data_2024 (1).json")
dc=pd.read_csv("old_airline_data_2023 (1).csv")
```

```python
print(dj.isnull().sum())
```

```python
dj.iloc[:,1:]=dj.iloc[:,1:].fillna(dj.iloc[:,1:].mean())
```

```python

dj=dj.dropna(subset=["Flights Operated","Passengers Carried","Target Passengers"], how='all')
```

```python
#dj["Month"]=dj["Month"].isin(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
```

```python

dj.columns=dj.columns.str.lower().str.replace(" ","_")
```

```python

dj["month"]=dj["month"].str.capitalize()
```

```python
dj.drop_duplicates(subset=["month"],inplace=True)
```

```python
print(dj)
```

```python
dc.columns=dc.columns.str.lower().str.replace(" ","_")
dc["month"]=dc["month"].str.capitalize()
```

```python
new = pd.concat([dc,dj],ignore_index=True)
print(new)
```

```python
new["Year"] = pd.Series([2023 if i < 53 else 2024 for i in new.index], index=new.index)
#new["Year"] = [2023 if i < len(dc) else 2024 for i in range(len(new))]
```

```python
mon=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
new = new[new['month'].isin(mon)]
print(new)
```

```python
new.sort_values(["Year","month"],inplace=True)
```

```python
new["month"]=pd.Categorical(new["month"], categories=mon, ordered=True)
print(new)
```

```python
new["pct_diff_target"]=((new["passengers_carried"]-new["target_passengers"])/new["target_passengers"])*100
print(new)
```

```python
target_not_achieved=new[new["pct_diff_target"] < 0]
print(target_not_achieved[["Year", "month", "pct_diff_target"]])
```

```python
total_passengers=new.groupby('Year')['passengers_carried'].sum()
total_flights=new.groupby('Year')['flights_operated'].sum()
Average_pc_diff_target=new.groupby('Year')['pct_diff_target'].mean()
annual_summary=pd.concat([total_passengers,total_flights,Average_pc_diff_target],ignore_index=True,axis=1)
print(annual_summary)
```

```python
best=new.loc[new["pct_diff_target"].idxmax()]
worst=new.loc[new["pct_diff_target"].idxmin()]
print(best,worst)
```

```python
new=new.to_csv("combined_airline_data.csv",index=False)
target_not_achieved=target_not_achieved.to_csv("target_not_achieved.csv",index=False)
annual_summary=annual_summary.to_json("yearly_summary.json",index=False)
```

```python
with zipfile.ZipFile("airline_analysis_outputs.zip","w") as zf:
    zf.write("combined_airline_data.csv")
    zf.write("target_not_achieved.csv")
    zf.write("yearly_summary.json")
```
